```python
import time
start = time.time()

from lasy.laser import Laser
from lasy.profiles.gaussian_profile import GaussianProfile
from lasy.optical_elements import Axiparabola

import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import c

import full_field
```

```python
l_w = 10.54e-7
f0 = 70e-2
delta = 2e-2
w = 10e-3
tau = 1.5e-14
E = 6.2
w0 = f0 * l_w / w / np.pi
vf = 1.0005*c
print(w/w0)
npoints = (8000, 2000)
print(npoints)
print("time:", (time.time()-start)/60, "min")
```

```python
profile = GaussianProfile(l_w, (1,0), E, w, tau, 0.0)
laser = Laser("rt", (0, -10*tau), (6*w, 6*tau), npoints, profile)
print("time:", (time.time()-start)/60, "min")
```

```python
axiparabola = Axiparabola(f0, delta, w)
import radialGroupDelay as RGD
def tau_D(r):
    return RGD.tau_D_const_v(r, vf, axiparabola)
radial_delay = RGD.RadialGroupDelay(tau_D, l_w)
laser.apply_optics(radial_delay)
laser.show()
print("time:", (time.time()-start)/60, "min")
```

```python
N = 14
M = 5

zs = np.zeros(M+1)
rs = np.zeros(N)
taus = np.zeros(N)
dts = np.zeros((M+1,N))


for n in range(N):
    print(n, "should be", tau_D(n*(2*w/(N-2))), "; is", full_field.get_tpeak(laser, nr=int(2*n*(npoints[0]/6/(N-2)))))
    rs[n] = n*(2*w/(N-2))
    taus[n] = tau_D(n*(2*w/(N-2)))
    dts[0, n] = full_field.get_tpeak(laser, nr=int(2*n*(npoints[0]/6/(N-2))))
for m in range(M):
    print("Step", m+1, "of", M, ":")
    laser.propagate(f0/3)
    laser.show()
    zs[m+1] = (m+1)*f0/3
    for n in range(N):
        print(n*(2*w/(N-2)), "should be", tau_D(n*(2*w/(N-2))), "; is", full_field.get_tpeak(laser, nr=int(2*n*(npoints[0]/6/(N-2)))))
        dts[m+1, n] = full_field.get_tpeak(laser, nr=int(2*n*(npoints[0]/6/(N-2))))
    
    print("time:", (time.time()-start)/60, "min")
```

```python
print(zs)
for n in range(N):
    s = str(rs[n])[:6]+"&"+str(taus[n])[:6]
    for m in range(M+1):
        s += "&"+str(dts[m,n]*1e15)[:6]
    s += "\\\\"
    print(s)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(np.concatenate((taus[:0:-1], taus))*1e15, np.concatenate((-rs[:0:-1], rs)), label="$\\tau_D(r)$")
for m in range(M+1):
    ax.plot(np.concatenate((dts[m,:0:-1], dts[m,:]))*1e15, np.concatenate((-rs[:0:-1], rs)), ".", label="$\\Delta t$ at $z="+str(zs[m])[:5]+"$ m")
ax.set_xlabel("$\\Delta t$/fs")
ax.set_ylabel("$r$/m")
ax.legend()
plt.show()
```
