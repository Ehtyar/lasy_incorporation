```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.array([(n+1)/3 for n in range(6)])
t1 = np.array([-0.0567,-0.1132,-0.1692,-0.2244,-0.2784,-0.3310])
t2 = np.array([-0.0272,-0.1079,-0.2413,-0.4263,-0.6620,-0.9474])
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(z, t2, label="theoretical t-z/c")
ax.plot(z, t1, ".", label="measured t-z/c")
ax.legend()
ax.set_xlabel("$z-f_0$/mm")
ax.set_ylabel("$t-z/c$/fs")
```

```python
z = np.array([(n+1)/3 for n in range(6)])
t1 = np.array([31.89,32.13,32.67,33.48,34.58,35.94])
t2 = np.array([33.84,23.93,19.54,16.92,15.13,13.81])
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(z, t2, label="theoretical w")
ax.plot(z, t1, ".", label="measured w")
ax.legend()
ax.set_xlabel("$z-f_0$/mm")
ax.set_ylabel("$w/\\mu$m")
```

```python
f0 = 7e-2
delta = 2e-3
R = 1.7e-3
r = np.linspace(-R, R, 400)
s_sm = r**2/(4*f0) - r**4*delta/(8*f0**2*R**2) + r**6*delta*(R**2+8*f0*delta)/(96*f0**4*R**4)
s_am = R**2/(4*delta) * np.log(1+delta/f0*(r/R)**2)
r2 = r**2
sag = (
            (1.0 / (4 * f0)) * r2
            - (delta / (8 * f0**2 * R**2)) * r2**2
            + delta
            * (R**2 + 8 * f0 * delta)
            / (96 * f0**4 * R**4)
            * r2**3
        )
s_pa = r**2/(4*f0)
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(r*1e3, s_pa*1e6, label="Parabola")
ax.plot(r*1e3, s_sm*1e6, label="Axiparabola Smartsev et al")
ax.plot(r*1e3, s_am*1e6, label="Axiparabola Ambat et al")
ax.plot(r*1e3, sag*1e6, label="Lasy implementation")
ax.legend()
ax.set_xlabel("$r$/mm")
ax.set_ylabel("$s_f(r)/\\mu$m")
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(r*1e3, s_sm*1e6-s_am*1e6, label="Axiparabola Smartsev et al - Axiparabola Ambat et al")
ax.plot(r*1e3, s_sm*1e6-sag*1e6, label="Axiparabola Smartsev et al - Lasy implementation")
#ax.plot(r*1e3, s_pa*1e6 - s_am*1e6, label="Axiparabola Smartsev et al")
ax.legend()
ax.set_xlabel("$r$/mm")
ax.set_ylabel("$\Delta s_f(r)/\\mu$m")
```
