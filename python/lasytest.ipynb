```python
gauss = profiles.GaussianProfile(4e-7, (1, 1), 1, 3e-4, 1e-14, 0)
laser = lasers.Laser("rt", (0, -2e-14), (5e-4, 2e-14), (5000, 100), gauss)
laser.show()
```

```python
f0 = 3e-2
delta = 1.5e-2
axiparabola = optics.Axiparabola(f0, delta, 4e-4)
laser.apply_optics(axiparabola)
laser.show()
```

```python
l=0
for n in range(3):
    laser.propagate(0.5*f0)
    laser.show()
    l += 0.5*f0
    print(l)

#for n in range(6):
#    laser.propagate(0.25*delta)
#    laser.show()
#    l += 0.25*delta
#    print(l)
```
