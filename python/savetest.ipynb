```python
from lasy.laser import Laser
from lasy.profiles.gaussian_profile import GaussianProfile
import numpy as np
```

```python
l_w = 10.54e-7
f0 = 70e-2
delta = 2e-2
w = 1e-5
tau = 1.5e-14
E = 6.2
w0 = f0 * l_w / w / np.pi
vf = 300000000
print(w/w0)
npoints = (11000, 3200)
print(npoints)
```

```python
profile = GaussianProfile(l_w, (1,0), E, w, tau, 0.0)
#propagator = AngularSpectrumPropagator(profile.omega0, "xyt")

laser = Laser("rt", (0, -7.*tau), (24.2*w, 7*tau), npoints, profile)
```

```python
import full_field
full_field.laser_to_openPMD(laser, "test_laser", Nt=1536, Nx=512, Ny=512, points_between_r=0.25, forced_dt=1.39e-16, data_step=2, conversion_safety=1.5)
```
