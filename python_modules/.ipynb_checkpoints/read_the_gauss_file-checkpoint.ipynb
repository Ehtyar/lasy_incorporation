```python
import ptime

import showdata
s1 = "/bigdata/hplsim/external/marqua27/"
s2 = "/bigdata/hplsim/production/marqua27/"
h = "/home/marqua27/jupyter_notebooks/diags/"
end = "/simOutput/openPMD/simData_%T.bp"
filename = s2+"vacuum_propagation/parabola01/run02"+end
#filename = h+"parabol.bp"
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import c
def shouldbe(t):
    """theoretical w"""
    l_w = 10.54e-7
    f0 = 7e-2
    w = 1e-3
    tau = 1.5e-14
    des_dt = 1.39e-16 # PIConGPU Standardwert
    w0 = f0 * l_w / w / np.pi
    zr = np.pi * w0**2/l_w
    z = c*t - c*39450*des_dt
    return w0 * np.sqrt(1+(z/zr)**2)
showdata.plot_w(filename=filename, forward="y", startit=1, unit=2.353377656651999e-05, unitname="$w_0$", comp_func=shouldbe)
ptime.ptime()
```
