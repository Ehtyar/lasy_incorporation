```python
from scipy.constants import c
import axiparabola_theory as axi
from lasy.optical_elements import Axiparabola
delta = 2e-3
f0 = 7e-2
w = 1e-3
R = 1.7*w
```

```python
l_w = 10.54e-7
w0 = f0 * l_w / w / np.pi
zr = np.pi * w0**2/l_w
def shouldbe(t):
    """theoretical w"""
    return axi.wtf(t+f0/c, Axiparabola(f0, delta, R), l_w)
def parab(t):
    return w0 * np.sqrt(1+(t*c/zr)**2)
showdata.plot_w(filename=filename, forward="y", startit=4000, comp_func=shouldbe, maxtime=axi.tfz((f0+delta) * (1-1e-7), Axiparabola(f0, delta, R))-f0/c)
```

```python
def shouldbe2(t):
    """theoretical value"""
    l_w = 10.54e-7
    tau = 1.5e-14
    return axi.ztf(t+f0/c, Axiparabola(f0, delta, R))-f0-c*t
showdata.plot_lpeak(filename=filename, forward="y", startit=4000, 
                    comp_func=shouldbe2, maxtime=axi.tfz((f0+delta) * (1-2e-7), Axiparabola(f0, delta, R))-f0/c,
                   apply_maxtime_to_measurement=True)
```
