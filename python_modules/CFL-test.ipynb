```python
from scipy.constants import c
import numpy as np

xi = 0.995
dx = 0.1772e-6
dy = 0.4430e-7
dz = dx

dt = 1.39e-16
```

```python
dt_cfl = xi / np.sqrt(1/dx**2 + 1/dy**2 + 1/dz**2) / c
print("dt should be:",dt_cfl)
print("error:", (dt/dt_cfl-1)*100, "%")
```
